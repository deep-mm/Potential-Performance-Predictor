```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Dataset/data.csv')
df
```

```python
df= df.drop(['EmployeeNumber', 'EmployeeCount'], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['BusinessTravel'] = le.fit_transform(df['BusinessTravel'])
df['Department'] = le.fit_transform(df['Department'])
df['EducationField'] = le.fit_transform(df['EducationField'])
df['Attrition'] = le.fit_transform(df['Attrition'])
df['Gender'] = le.fit_transform(df['Gender'])
df['JobRole'] = le.fit_transform(df['JobRole'])
df['MaritalStatus'] = le.fit_transform(df['MaritalStatus'])
df['OverTime'] = le.fit_transform(df['OverTime'])
df['Over18'] = le.fit_transform(df['Over18'])
df['JobSatisfaction'] = le.fit_transform(df['JobSatisfaction'])
df['EnvironmentSatisfaction'] = le.fit_transform(df['EnvironmentSatisfaction'])
df
```

```python
#total Nan values of all columns
df.isnull().sum().sum()
```

```python
#All cloumns names
df.columns
```

```python
Y_all = df['Attrition']
X_all = df.drop(['Attrition'], axis=1)
```

```python
#plot graph with Y values
un, cn = np.unique(Y_all, return_counts=True)
plt.bar(un, cn)
plt.show()
un, cn
```

```python
#SMOTE
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=2)

X_res, Y_res = sm.fit_resample(X_all, Y_all.ravel())

#plot graph with Y_res values
un, cn = np.unique(Y_res, return_counts=True)
plt.bar(un, cn)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import random

def get_trained_model(X, Y, model):
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=111)
    model.fit(X_train, Y_train)
    Y_pred = model.predict(X_test)
    print('Accuracy Score: ', accuracy_score(Y_test, Y_pred))
    print('Confusion Matrix: \n', confusion_matrix(Y_test, Y_pred))
    print('Classification Report: \n', classification_report(Y_test, Y_pred))
    return model

```

```python
#Random Forest with all selected features
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, random_state=0)
model = get_trained_model(X_res, Y_res, model)
```

```python
#Feature Selection with SelectKBest with select 10 best features names
from sklearn.feature_selection import SelectKBest, chi2
bestfeatures = SelectKBest(score_func=chi2, k=10)
fit = bestfeatures.fit(X_res,Y_res)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X_res.columns)
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']
print(featureScores.sort_values(by=['Score'], ascending=False))

cols_KBest = featureScores.nlargest(20,'Score')['Specs'].values
```

```python
X_KBest = X_res[cols_KBest]
model = RandomForestClassifier(n_estimators=100, random_state=0)
model = get_trained_model(X_KBest, Y_res, model)
```

```python
#Feature Selection using ExtraTreesClassifier
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
model = ExtraTreesClassifier()
model.fit(X_res, Y_res)
print(model.feature_importances_)
feat_importances = pd.Series(model.feature_importances_, index=X_all.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.show()

cols_Tree = feat_importances.nlargest(20).index
```

```python
X_ExtraTrees = X_res[cols_Tree]
model = RandomForestClassifier(n_estimators=100, random_state=0)
model = get_trained_model(X_ExtraTrees, Y_res, model)
```

```python
#Feature selection using lasso regression
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
feature_sel_model = SelectFromModel(Lasso(alpha=0.005, random_state=0))
feature_sel_model.fit(X_res, Y_res)
feature_sel_model.get_support()
selected_feat = X_res.columns[(feature_sel_model.get_support())]
print('total features: {}'.format((X_res.shape[1])))
print('selected features: {}'.format(len(selected_feat)))
print(selected_feat)
print('features with coefficients shrank to zero: {}'.format(np.sum(feature_sel_model.estimator_.coef_ == 0)))
deleted_feat = X_res.columns[(feature_sel_model.estimator_.coef_ == 0).ravel().tolist()]
print(deleted_feat)

cols_lasso = selected_feat
```

```python
X_Lasso = X_res[cols_lasso]
model = RandomForestClassifier(n_estimators=100, random_state=0)
model = get_trained_model(X_Lasso, Y_res, model)
```

```python
#Feature selection using PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=5)
pca.fit(X_res)
X_pca = pca.transform(X_res)
print(X_pca.shape)
print(pca.explained_variance_ratio_)


plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()
```

```python
model = RandomForestClassifier(n_estimators=100, random_state=0)
model = get_trained_model(X_pca, Y_res, model)
```
