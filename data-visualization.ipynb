```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Dataset/data.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['BusinessTravel'] = le.fit_transform(df['BusinessTravel'])
df['Department'] = le.fit_transform(df['Department'])
df['EducationField'] = le.fit_transform(df['EducationField'])
df['Attrition'] = le.fit_transform(df['Attrition'])
df['Gender'] = le.fit_transform(df['Gender'])
df['JobRole'] = le.fit_transform(df['JobRole'])
df['MaritalStatus'] = le.fit_transform(df['MaritalStatus'])
df['OverTime'] = le.fit_transform(df['OverTime'])
df['Over18'] = le.fit_transform(df['Over18'])
df['JobSatisfaction'] = le.fit_transform(df['JobSatisfaction'])
df['EnvironmentSatisfaction'] = le.fit_transform(df['EnvironmentSatisfaction'])
```

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

sel = SelectFromModel(Lasso(alpha=0.005, random_state=42))
sel.fit(X, Y_JS)
print("Job Statisfaction", sel.get_support())
print("Selected Features", X.columns[sel.get_support()])
X_JS = X[X.columns[sel.get_support()]]

sel = SelectFromModel(Lasso(alpha=0.005, random_state=42))
sel.fit(X, Y_ES)
print("\n\nEmployee Statisfaction", sel.get_support())
print("Selected Features", X.columns[sel.get_support()])
X_ES = X[X.columns[sel.get_support()]]


sel = SelectFromModel(Lasso(alpha=0.005, random_state=42))
sel.fit(X, Y_A)
print("\n\nAttrition", sel.get_support())
print("Selected Features", X.columns[sel.get_support()])
X_A = X[X.columns[sel.get_support()]]

```

```python
from sklearn.model_selection import train_test_split
X_JS_train, X_JS_test, Y_JS_train, Y_JS_test = train_test_split(X_JS, Y_JS, test_size=0.2, random_state=42, stratify=Y_JS)
X_ES_train, X_ES_test, Y_ES_train, Y_ES_test = train_test_split(X_ES, Y_ES, test_size=0.2, random_state=42, stratify=Y_ES)
X_A_train, X_A_test, Y_A_train, Y_A_test = train_test_split(X_A, Y_A, test_size=0.2, random_state=42, stratify=Y_A)
```

```python
X_JS_train.to_csv('Dataset/X_JS_train.csv', index=False)
X_JS_test.to_csv('Dataset/X_JS_test.csv', index=False)
Y_JS_train.to_csv('Dataset/Y_JS_train.csv', index=False)
Y_JS_test.to_csv('Dataset/Y_JS_test.csv', index=False)

X_ES_train.to_csv('Dataset/X_ES_train.csv', index=False)
X_ES_test.to_csv('Dataset/X_ES_test.csv', index=False)
Y_ES_train.to_csv('Dataset/Y_ES_train.csv', index=False)
Y_ES_test.to_csv('Dataset/Y_ES_test.csv', index=False)


X_A_train.to_csv('Dataset/X_A_train.csv', index=False)
X_A_test.to_csv('Dataset/X_A_test.csv', index=False)
Y_A_train.to_csv('Dataset/Y_A_train.csv', index=False)
Y_A_test.to_csv('Dataset/Y_A_test.csv', index=False)
```

```python
#write a function to create a model for each of the three target variables using pipeline
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

def create_model(model, X_train, Y_train, X_test, Y_test, Y_name, name):
    model.fit(X_train, Y_train)
    Y_pred = model.predict(X_test)
    print("\n\n\n" + Y_name + " " + name + " Model")
    print("Accuracy: ", accuracy_score(Y_test, Y_pred))
    print("Precision: ", precision_score(Y_test, Y_pred, average='macro'))
    print("Recall: ", recall_score(Y_test, Y_pred, average='macro'))
    print("F1 Score: ", f1_score(Y_test, Y_pred, average='macro'))
    print("Confusion Matrix: \n", confusion_matrix(Y_test, Y_pred))
    print("Classification Report: \n", classification_report(Y_test, Y_pred))
```

```python
def get_models(model, name):    
    #Job Satisfaction
    model_JS = Pipeline([('scaler', StandardScaler()), ('model', model)])
    create_model(model_JS, X_JS_train, Y_JS_train, X_JS_test, Y_JS_test, "JS", name)

    #Environment Satisfaction
    model_ES = Pipeline([('scaler', StandardScaler()), ('model', model)])
    create_model(model_ES, X_ES_train, Y_ES_train, X_ES_test, Y_ES_test, "ES", name)

    #Attrition
    model_A = Pipeline([('scaler', StandardScaler()), ('model', model)])
    create_model(model_A, X_A_train, Y_A_train, X_A_test, Y_A_test, "A", name)

    return model_JS, model_ES, model_A
```

```python
#Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
model_JS_RF, model_ES_RF, model_A_RF = get_models(RandomForestClassifier(n_estimators=200, random_state=42), "Random Forest")
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
model_JS_LR, model_ES_LR, model_A_LR = get_models(LogisticRegression(random_state=42), "Logistic Regression")
```

```python

#AdaBoost Classifier
from sklearn.ensemble import AdaBoostClassifier
model_JS_AB, model_ES_AB, model_A_AB = get_models(AdaBoostClassifier(n_estimators=200, random_state=42), "AdaBoost")
```

```python

#XGBoost Classifier for multi-class classification
from xgboost import XGBClassifier
model_JS_XGB, model_ES_XGB, model_A_XGB = get_models(XGBClassifier(n_estimators=200, random_state=42), "XGBoost")
```
